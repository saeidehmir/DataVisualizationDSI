```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
          y,
          marker = '>',
           color = "indigo")
#fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,
        y,
        marker = '*',
        color = "indigo",
        linestyle = '--',
        linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y,
        marker = 'D',
        markersize = 12,
        color = '#7425B9',
        linestyle = '--',
        linewidth =2,
        markeredgecolor = '#FA9359',
        markerfacecolor = '#000000')

ax.grid(axis = 'y',
       color = 'dodgerblue',
       linewidth =2,
       linestyle = 'dashdot')
```
