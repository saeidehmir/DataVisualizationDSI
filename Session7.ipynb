```python
#!pip install matplotlib-venn

from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
A = set(["apple","banana","watermelon"])
B = set(["pumpkin","blueberry","apple","key lime"])

diagram = venn2_unweighted([A,B],
                          set_labels = ('Fruits','Pies'),
                          set_colors = ("blue","red"),
                          alpha = 0.5 # transparency 
)   

diagram.get_label_by_id("10").set_text("\n".join(A-B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B-A))

```

```python
#!pip install plottable

from plottable import Table
from plottable import ColumnDefinition
import textwrap as twp
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/IvanJoskic/pet-data-set/main/pet-set.csv",
                on_bad_lines = 'skip',
                usecols = ['breedname', 'weight','gender'])

df = df.set_index("breedname")

df

```

```python
fig, ax = plt.subplots(figsize = (15,5))
ax.tab = Table(df)
```

```python
column_names = [ColumnDefinition(name = "breedname",
                                 title = "Breed Name",
                                 textprops = {"ha":"center","weight":"bold","size":14}),
                ColumnDefinition(name = "weight",
                                 title = "Weight",
                                 textprops = {"ha":"center",
                                             "bbox":{"boxstyle":"circle","pad":0.35, "color":"skyblue"}}),
                ColumnDefinition(name = "gender",
                                 title = "Gender",
                                 textprops = {"ha":"center"})
    
]

fig, ax = plt.subplots(figsize = (15,5))

ax.tab = Table(df,
              column_definitions = column_names,
              row_divider_kw = {"linewidth":1, "linestyle":"dashed"})


```
